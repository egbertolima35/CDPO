```python
import pandas as pd
from google.colab import files
import numpy as np
```

```python
uploaded = files.upload()
```

```python
for filename in uploaded.keys():
  print(f'Arquivo "{filename}" carregado com sucesso.')


```

```python
df = pd.read_excel(filename)
```

```python
display(df.head())
```

```python
candidatos = pd.DataFrame(df)
```

```python
candidatos = candidatos.rename(columns={
    'Nome completo:': 'Nome',
    'Cargo:': 'Cargo',
    'Local/Trabalho:': 'Local'
})
```

```python
selecionados = candidatos[['Nome', 'Cargo', 'Local']]
```

```python
novo_banco = candidatos[['Nome', 'Cargo', 'Local']]
```

```python
display(selecionados)
```

```python
novo_banco.to_excel('eleitos.xlsx', index=False)  # ou .to_csv('eleitos.csv', index=False)
```
